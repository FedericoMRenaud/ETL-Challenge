```python
%matplotlib notebook

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Importing data files for Starbucks
starbucks_file = "starbucks_data.csv"
starbucks_df = pd.read_csv(starbucks_file)
starbucks_df.head()
```

```python
# Data Cleanup
starbucks_df = starbucks_df.replace("-","")
starbucks_df.head()
```

```python
# Data Cleanup
starbucks_df[starbucks_df.Calories!='']
starbucks_df.head()
```

```python
# Data Cleanup
starbucks_df.columns = ['Item','Company','Calories','Fat_g','Carb_g','Fiber_g','Protein','Sodium']
starbucks_df.head()
```

```python
# Data Cleanup
starbucks_final = starbucks_df[starbucks_df.Calories!='']
starbucks_final
```

```python
# Importing data files McDonalds
mcdonalds_file = "mcdonalds_data.csv"
mcdonalds_df = pd.read_csv(mcdonalds_file)
mcdonalds_df.head()
```

```python
# Data Cleanup
mcdonalds_clean = mcdonalds_df.drop(columns=["Category",
                                           "Serving Size","Calories from Fat","Total Fat (% Daily Value)",
                                           "Saturated Fat","Saturated Fat (% Daily Value)", 
                                           "Trans Fat", "Carbohydrates (% Daily Value)",
                                           "Dietary Fiber (% Daily Value)", "Sugars", 
                                           "Vitamin A (% Daily Value)","Vitamin C (% Daily Value)",
                                           "Calcium (% Daily Value)","Iron (% Daily Value)", 
                                           "Cholesterol (% Daily Value)","Cholesterol",
                                           "Sodium (% Daily Value)"])
mcdonalds_clean.head()
```

```python
# Data Cleanup
mcdonalds_final = mcdonalds_clean.rename(columns={"Total Fat":"Fat_g","Carbohydrates":"Carb_g",
                                                "Dietary Fiber":"Fiber_g"})
mcdonalds_final.head()
```

```python
# Data exploration for analysis
med_cal_st = starbucks_final["Calories"].median()
med_cal_mc = mcdonalds_final["Calories"].median()
print(f'Median Calories - Starbucks = {med_cal_st}')
print(f'Median Calories - McDonalds = {med_cal_mc}')
```

```python
# Data exploration for analysis
min_cal_st = starbucks_final["Calories"].min()
min_cal_mc = mcdonalds_final["Calories"].min()
print(f'Min. Calories - Starbucks = {min_cal_st}')
print(f'Min. Calories - McDonalds = {min_cal_mc}')
```

```python
# Data exploration for analysis
max_cal_st = starbucks_final["Calories"].max()
max_cal_mc = mcdonalds_final["Calories"].max()
print(f'Max. Calories - Starbucks = {max_cal_st}')
print(f'Max. Calories - McDonalds = {max_cal_mc}')
```

```python
# Data exploration for analysis
max_fat_st = starbucks_final["Fat_g"].max()
max_fat_mc = mcdonalds_final["Fat_g"].max()
print(f'Max. Fat - Starbucks = {max_fat_st}')
print(f'Max. Fat - McDonalds = {max_fat_mc}')
```

```python
# Merging dataframes
merged_df = mcdonalds_final.append(starbucks_final, sort=True)
merged_df.head()
```

```python
# Re-ordering columns 
merged_df_r = merged_df[["Company","Item","Calories","Carb_g","Fat_g","Fiber_g","Protein","Sodium"]]
merged_df_r.head()
```

```python
# To save final version of df to csv
merged_df_r.to_csv('md_vs_st.csv', index=False) 
print('Dataframe to CSV created')
```

```python
print(merged_df_r.dtypes)
```

```python
# Convert variables into floats
merged_df_r['Calories'] = merged_df_r['Calories'].astype(float)
merged_df_r['Carb_g'] = merged_df_r['Carb_g'].astype(float)
merged_df_r['Fat_g'] = merged_df_r['Fat_g'].astype(float)
merged_df_r['Fiber_g'] = merged_df_r['Fiber_g'].astype(float)
merged_df_r['Protein'] = merged_df_r['Protein'].astype(float)
merged_df_r['Sodium'] = merged_df_r['Sodium'].astype(float)
print(merged_df_r.dtypes)
```

```python
# Calories
calories_comp  = merged_df_r.groupby(['Company'])
calories_comp.Calories.mean()
```

```python
# Fat
fat_comp  = merged_df_r.groupby(['Company'])
fat_comp.Fat_g.mean()
```
